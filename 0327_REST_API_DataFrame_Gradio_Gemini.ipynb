# **4.네이버와 구글의 서버 헤더 정보 비교**

---



```python
import requests

# 네이버 서버 헤더
naver_response = requests.get("https://www.naver.com")
print("=== 네이버 서버 헤더 ===")
for key, value in naver_response.headers.items():
    print(f"{key}: {value}")

print("\n")

# 구글 서버 헤더
google_response = requests.get("https://www.google.com")
print("=== 구글 서버 헤더 ===")
for key, value in google_response.headers.items():
    print(f"{key}: {value}")

# 주요 항목 비교
print("\n=== 주요 헤더 비교 ===")
compare_keys = ['Server', 'Content-Type', 'Content-Encoding']
for key in compare_keys:
    naver_val = naver_response.headers.get(key, "없음")
    google_val = google_response.headers.get(key, "없음")
    print(f"\n[{key}]")
    print(f"  네이버: {naver_val}")
    print(f"  구글:   {google_val}")
```

# **7.Gradio UI로 실시간 필터링**

---



```python
import requests
import pandas as pd
import gradio as gr

def filter_by_userid(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {"userId": user_id}
    response = requests.get(url, params=params)
    data = response.json()

    if not data:
        return pd.DataFrame({"결과": ["해당 userId의 데이터가 없습니다."]})

    df = pd.DataFrame(data)
    return df

demo = gr.Interface(
    fn=filter_by_userid,
    inputs=gr.Number(label="userId 입력 (1~10)", precision=0),
    outputs=gr.Dataframe(label="수집한 데이터 (Pandas DataFrame)"),
    title="네트워크 & 웹 데이터 수집 실습",
    description="userId를 입력하면 해당 사용자의 데이터를 Pandas 데이터프레임으로 가져옵니다."
)
print("=" * 60)
print("📝 Gradio UI가 실행되었습니다!")

print("=" * 60)
demo.launch(share=False)
```

# **8.서울시 공공자전거 이용정보 통계분석 대시보드 (2025.10~12)**

```python
###############################################################################
# 서울시 공공자전거 이용정보 통계분석 (Gradio UI)
# - 2025년 10월, 11월, 12월 데이터 merge 후 다양한 통계분석
# - Google Colab에서 실행 (Google Drive 연동)
###############################################################################

# ============================================================
# [셀 1] 필요한 라이브러리 설치
# ============================================================
# !pip install gradio

# ============================================================
# [셀 2] Google Drive 마운트
# ============================================================
from google.colab import drive
drive.mount('/content/drive')
print("✅ Google Drive 마운트 완료!")

# ============================================================
# [셀 3] 라이브러리 import
# ============================================================
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')

# Colab 한글 폰트 설정
import matplotlib.font_manager as fm
import os

os.system('apt-get install -y fonts-nanum > /dev/null 2>&1')
fm._load_fontmanager(try_read_cache=False)
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['axes.unicode_minus'] = False

# ============================================================
# [셀 4] CSV 파일 읽기 & 데이터 병합
# ============================================================
# Google Drive 경로 (내 드라이브 > Colab Notebooks)
folder = '/content/drive/MyDrive/Colab Notebooks/'

file_names = [
    '서울특별시 공공자전거 이용정보(일별)_2510.csv',
    '서울특별시 공공자전거 이용정보(일별)_2511.csv',
    '서울특별시 공공자전거 이용정보(일별)_2512.csv'
]

df_list = []
for fname in file_names:
    filepath = folder + fname
    try:
        temp = pd.read_csv(filepath, encoding='cp949')
    except UnicodeDecodeError:
        temp = pd.read_csv(filepath, encoding='utf-8')
    df_list.append(temp)
    print(f"✅ {fname}: {len(temp):,}행 로드 완료")

df = pd.concat(df_list, ignore_index=True)
print(f"\n📊 전체 병합 데이터: {len(df):,}행 x {len(df.columns)}열")
print(f"📋 컬럼: {df.columns.tolist()}")

# ============================================================
# [셀 5] 데이터 전처리
# ============================================================
# 컬럼명 앞뒤 공백 제거
df.columns = df.columns.str.strip()

# 대여일자를 datetime으로 변환
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['월'] = df['대여일자'].dt.month
df['요일'] = df['대여일자'].dt.day_name()
df['일'] = df['대여일자'].dt.day

# 요일 한글 매핑
day_map = {
    'Monday': '월요일', 'Tuesday': '화요일', 'Wednesday': '수요일',
    'Thursday': '목요일', 'Friday': '금요일', 'Saturday': '토요일', 'Sunday': '일요일'
}
df['요일_한글'] = df['요일'].map(day_map)

# 수치형 컬럼 확인 및 변환
numeric_cols = ['이용건수', '운동량', '탄소량', '이동거리(M)', '이용시간(분)']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("✅ 데이터 전처리 완료")
print(f"📅 기간: {df['대여일자'].min().date()} ~ {df['대여일자'].max().date()}")
df.head()

# ============================================================
# [셀 6] Gradio UI 구성 및 실행
# ============================================================
import gradio as gr

# ---- 탭1: 기본 통계 요약 ----
def basic_stats():
    stats = df[numeric_cols].describe().round(2).reset_index()
    stats.columns = ['통계항목'] + numeric_cols
    return stats

def data_overview():
    info = {
        '항목': ['전체 데이터 수', '기간', '대여소 수', '월별 데이터'],
        '값': [
            f"{len(df):,}행",
            f"{df['대여일자'].min().date()} ~ {df['대여일자'].max().date()}",
            f"{df['대여소'].nunique():,}개",
            ', '.join([f"{m}월: {cnt:,}건" for m, cnt in df.groupby('월').size().items()])
        ]
    }
    return pd.DataFrame(info)

# ---- 탭2: 일별 이용 추이 ----
def daily_trend():
    daily = df.groupby('대여일자')['이용건수'].sum().reset_index()
    daily = daily.sort_values('대여일자')

    fig, ax = plt.subplots(figsize=(14, 5))
    ax.plot(daily['대여일자'], daily['이용건수'], color='#2196F3', linewidth=1.5)
    ax.fill_between(daily['대여일자'], daily['이용건수'], alpha=0.15, color='#2196F3')
    ax.set_title('일별 공공자전거 이용건수 추이', fontsize=14, fontweight='bold')
    ax.set_xlabel('날짜')
    ax.set_ylabel('이용건수')
    ax.grid(True, alpha=0.3)
    fig.autofmt_xdate()
    plt.tight_layout()
    return fig

# ---- 탭3: 월별 비교 ----
def monthly_comparison():
    monthly = df.groupby('월')[numeric_cols].sum().reset_index()

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    colors = ['#4CAF50', '#2196F3', '#FF9800']
    month_labels = [f'{m}월' for m in monthly['월']]

    # 이용건수
    axes[0, 0].bar(month_labels, monthly['이용건수'], color=colors)
    axes[0, 0].set_title('월별 이용건수', fontweight='bold')
    for i, v in enumerate(monthly['이용건수']):
        axes[0, 0].text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

    # 이동거리
    axes[0, 1].bar(month_labels, monthly['이동거리(M)'] / 1000, color=colors)
    axes[0, 1].set_title('월별 총 이동거리 (km)', fontweight='bold')
    for i, v in enumerate(monthly['이동거리(M)'] / 1000):
        axes[0, 1].text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

    # 이용시간
    axes[1, 0].bar(month_labels, monthly['이용시간(분)'] / 60, color=colors)
    axes[1, 0].set_title('월별 총 이용시간 (시간)', fontweight='bold')
    for i, v in enumerate(monthly['이용시간(분)'] / 60):
        axes[1, 0].text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

    # 탄소량
    axes[1, 1].bar(month_labels, monthly['탄소량'], color=colors)
    axes[1, 1].set_title('월별 탄소 절감량', fontweight='bold')
    for i, v in enumerate(monthly['탄소량']):
        axes[1, 1].text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

    plt.suptitle('월별 주요 지표 비교 (2025년 10~12월)', fontsize=15, fontweight='bold', y=1.02)
    plt.tight_layout()
    return fig

# ---- 탭4: 성별 분석 ----
def gender_analysis():
    gender = df.groupby('성별')['이용건수'].sum().reset_index()
    gender = gender[gender['성별'].isin(['M', 'F', '남', '여', '남성', '여성'])]

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].pie(gender['이용건수'], labels=gender['성별'], autopct='%1.1f%%',
                colors=['#42A5F5', '#EF5350'], startangle=90, textprops={'fontsize': 13})
    axes[0].set_title('성별 이용 비율', fontweight='bold')

    axes[1].bar(gender['성별'], gender['이용건수'], color=['#42A5F5', '#EF5350'])
    axes[1].set_title('성별 이용건수', fontweight='bold')
    axes[1].set_ylabel('이용건수')
    for i, v in enumerate(gender['이용건수']):
        axes[1].text(i, v, f'{v:,.0f}', ha='center', va='bottom')

    plt.tight_layout()
    return fig

# ---- 탭5: 연령대별 분석 ----
def age_analysis():
    age = df.groupby('연령대')['이용건수'].sum().reset_index()
    age = age.sort_values('이용건수', ascending=False)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    bars = axes[0].barh(age['연령대'].astype(str), age['이용건수'], color='#7E57C2')
    axes[0].set_title('연령대별 이용건수', fontweight='bold')
    axes[0].set_xlabel('이용건수')
    for bar in bars:
        w = bar.get_width()
        axes[0].text(w, bar.get_y() + bar.get_height()/2, f'{w:,.0f}',
                     ha='left', va='center', fontsize=9)

    age_month = df.groupby(['월', '연령대'])['이용건수'].sum().reset_index()
    pivot = age_month.pivot(index='연령대', columns='월', values='이용건수').fillna(0)
    pivot.plot(kind='bar', ax=axes[1], color=['#4CAF50', '#2196F3', '#FF9800'])
    axes[1].set_title('월별 x 연령대 이용건수', fontweight='bold')
    axes[1].set_ylabel('이용건수')
    axes[1].legend(title='월', labels=[f'{c}월' for c in pivot.columns])
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    return fig

# ---- 탭6: 인기 대여소 TOP N ----
def top_stations(n):
    n = int(n)
    top = df.groupby('대여소')['이용건수'].sum().nlargest(n).reset_index()
    top.columns = ['대여소', '총 이용건수']
    top['순위'] = range(1, len(top) + 1)
    top = top[['순위', '대여소', '총 이용건수']]
    return top

def top_stations_chart(n):
    n = int(n)
    top = df.groupby('대여소')['이용건수'].sum().nlargest(n)

    fig, ax = plt.subplots(figsize=(12, max(5, n * 0.4)))
    colors = plt.cm.Blues([(i + 3) / (n + 3) for i in range(n)])[::-1]
    bars = ax.barh(range(n), top.values, color=colors)
    ax.set_yticks(range(n))
    ax.set_yticklabels(top.index, fontsize=9)
    ax.set_title(f'이용건수 TOP {n} 대여소', fontsize=14, fontweight='bold')
    ax.set_xlabel('총 이용건수')
    for bar in bars:
        w = bar.get_width()
        ax.text(w, bar.get_y() + bar.get_height()/2, f'{w:,.0f}',
                ha='left', va='center', fontsize=9)
    ax.invert_yaxis()
    plt.tight_layout()
    return fig

# ---- 탭7: 요일별 분석 ----
def weekday_analysis():
    day_order = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
    weekday = df.groupby('요일_한글')['이용건수'].sum().reindex(day_order)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    colors = ['#78909C'] * 5 + ['#EF5350', '#EF5350']
    axes[0].bar(weekday.index, weekday.values, color=colors)
    axes[0].set_title('요일별 이용건수', fontweight='bold')
    axes[0].set_ylabel('이용건수')
    axes[0].tick_params(axis='x', rotation=45)
    for i, v in enumerate(weekday.values):
        axes[0].text(i, v, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

    weekday_dist = df.groupby('요일_한글')['이동거리(M)'].mean().reindex(day_order) / 1000
    axes[1].bar(weekday_dist.index, weekday_dist.values, color=colors)
    axes[1].set_title('요일별 평균 이동거리 (km)', fontweight='bold')
    axes[1].set_ylabel('평균 이동거리 (km)')
    axes[1].tick_params(axis='x', rotation=45)
    for i, v in enumerate(weekday_dist.values):
        axes[1].text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=9)

    plt.tight_layout()
    return fig

# ---- 탭8: 이동거리·이용시간 분포 ----
def distribution_analysis():
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    dist_data = df['이동거리(M)'].dropna()
    dist_data = dist_data[dist_data > 0]
    axes[0].hist(dist_data / 1000, bins=50, color='#26A69A', edgecolor='white', alpha=0.8)
    axes[0].set_title('이동거리 분포', fontweight='bold')
    axes[0].set_xlabel('이동거리 (km)')
    axes[0].set_ylabel('빈도')
    axes[0].axvline(dist_data.mean() / 1000, color='red', linestyle='--',
                    label=f'평균: {dist_data.mean()/1000:.2f}km')
    axes[0].legend()

    time_data = df['이용시간(분)'].dropna()
    time_data = time_data[time_data > 0]
    axes[1].hist(time_data, bins=50, color='#FFA726', edgecolor='white', alpha=0.8)
    axes[1].set_title('이용시간 분포', fontweight='bold')
    axes[1].set_xlabel('이용시간 (분)')
    axes[1].set_ylabel('빈도')
    axes[1].axvline(time_data.mean(), color='red', linestyle='--',
                    label=f'평균: {time_data.mean():.1f}분')
    axes[1].legend()

    plt.tight_layout()
    return fig

# ============================================================
# Gradio Blocks UI 구성
# ============================================================
with gr.Blocks(
    title="서울시 공공자전거 통계분석",
    theme=gr.themes.Soft()
) as demo:

    gr.Markdown("""
    # 🚲 서울시 공공자전거(따릉이) 이용정보 통계분석
    ### 2025년 10월 ~ 12월 (3개월) 일별 데이터 분석
    ---
    """)

    with gr.Tab("📋 데이터 개요"):
        gr.Markdown("### 데이터 기본 정보")
        overview_btn = gr.Button("📊 개요 보기", variant="primary")
        overview_output = gr.Dataframe(label="데이터 개요")
        overview_btn.click(fn=data_overview, outputs=overview_output)

        gr.Markdown("### 수치형 데이터 기본 통계")
        stats_btn = gr.Button("📈 통계 보기", variant="primary")
        stats_output = gr.Dataframe(label="기술 통계량")
        stats_btn.click(fn=basic_stats, outputs=stats_output)

    with gr.Tab("📈 일별 이용 추이"):
        gr.Markdown("### 일별 공공자전거 이용건수 변화")
        daily_btn = gr.Button("그래프 보기", variant="primary")
        daily_output = gr.Plot()
        daily_btn.click(fn=daily_trend, outputs=daily_output)

    with gr.Tab("📊 월별 비교"):
        gr.Markdown("### 10월 vs 11월 vs 12월 주요 지표 비교")
        monthly_btn = gr.Button("그래프 보기", variant="primary")
        monthly_output = gr.Plot()
        monthly_btn.click(fn=monthly_comparison, outputs=monthly_output)

    with gr.Tab("👫 성별 분석"):
        gr.Markdown("### 성별 이용 현황")
        gender_btn = gr.Button("그래프 보기", variant="primary")
        gender_output = gr.Plot()
        gender_btn.click(fn=gender_analysis, outputs=gender_output)

    with gr.Tab("👥 연령대별 분석"):
        gr.Markdown("### 연령대별 이용 현황")
        age_btn = gr.Button("그래프 보기", variant="primary")
        age_output = gr.Plot()
        age_btn.click(fn=age_analysis, outputs=age_output)

    with gr.Tab("🏆 인기 대여소"):
        gr.Markdown("### 이용건수 기준 인기 대여소 순위")
        n_input = gr.Slider(minimum=5, maximum=30, value=10, step=1, label="TOP N")
        station_btn = gr.Button("조회", variant="primary")
        station_table = gr.Dataframe(label="인기 대여소 순위")
        station_chart = gr.Plot()
        station_btn.click(fn=top_stations, inputs=n_input, outputs=station_table)
        station_btn.click(fn=top_stations_chart, inputs=n_input, outputs=station_chart)

    with gr.Tab("📅 요일별 분석"):
        gr.Markdown("### 요일별 이용 패턴 (주중 vs 주말)")
        weekday_btn = gr.Button("그래프 보기", variant="primary")
        weekday_output = gr.Plot()
        weekday_btn.click(fn=weekday_analysis, outputs=weekday_output)

    with gr.Tab("📉 분포 분석"):
        gr.Markdown("### 이동거리 & 이용시간 분포")
        dist_btn = gr.Button("그래프 보기", variant="primary")
        dist_output = gr.Plot()
        dist_btn.click(fn=distribution_analysis, outputs=dist_output)

    gr.Markdown("""
    ---
    *데이터 출처: [서울 열린데이터광장](https://data.seoul.go.kr) - 서울특별시 공공자전거 이용정보(일별)*
    """)
print("=" * 60)
print("🚲 Gradio UI가 실행되었습니다!")
print("=" * 60)
demo.launch(share=False)
```

# **9. Agent AI - Gemini 번역 서비스**

---




```python

import requests
import pandas as pd
import gradio as gr
from google.colab import ai

# ============================================================
# [셀 3] Gradio UI 구성 및 실행
# ============================================================

def get_posts(user_id):
    """userId로 posts 데이터 가져오기 (7번 과제와 동일)"""
    url = "https://jsonplaceholder.typicode.com/posts"
    params = {"userId": int(user_id)}
    response = requests.get(url, params=params)
    data = response.json()

    if not data:
        return pd.DataFrame({"결과": ["해당 userId의 데이터가 없습니다."]}), "데이터가 없습니다."

    df = pd.DataFrame(data)

    # 첫 1행만 번역 (수업 노트 지시사항)
    first_row = data[0]
    text_to_translate = f"Title: {first_row['title']}\nBody: {first_row['body']}"

    prompt = f"다음 영문 텍스트를 한국어로 자연스럽게 번역해 주세요:\n\n{text_to_translate}"
    translated = ai.generate_text(prompt)

    return df, translated

# Gradio UI
demo = gr.Interface(
    fn=get_posts,
    inputs=gr.Number(label="userId 입력 (1~10)", precision=0, value=1),
    outputs=[
        gr.Dataframe(label="수집한 데이터 (Pandas DataFrame)"),
        gr.Textbox(label="첫 번째 행 한국어 번역 (Gemini)")
    ],
    title="📝 Posts 데이터 번역 서비스 (Gemini AI)",
    description="userId를 입력하면 데이터를 가져오고, 첫 번째 행을 Gemini AI가 한국어로 번역합니다.\n(google.colab.ai 사용 - API 키 불필요)"
)

print("=" * 60)
print("📝 Gradio UI가 실행되었습니다!")
print("아래 링크를 클릭하면 번역 서비스를 확인할 수 있습니다.")
print("=" * 60)

demo.launch(share=False)
```

```python
from google import genai
from google.genai import types

client = genai.Client(api_key=GEMINI_API_KEY)
```

```python
MODEL_ID = "gemini-3.1-pro-preview" # @param ["gemini-2.5-flash-lite", "gemini-2.5-flash", "gemini-2.5-pro", "gemini-3.1-flash-lite-preview", "gemini-3-flash-preview", "gemini-3.1-pro-preview"] {"allow-input":true, isTemplate: true}
```

```python
import gradio as gr

def generate_text_with_gemini(prompt):
    try:
        response = client.models.generate_content(
            model=MODEL_ID,
            contents=prompt
        )
        return response.text
    except Exception as e:
        return f"An error occurred: {e}"


iface = gr.Interface(
    fn=generate_text_with_gemini,
    inputs=gr.Textbox(lines=5, label="Enter your text prompt:"),
    outputs=gr.Textbox(label="Generated Text:"),
    title="Gemini Text Generation App",
    description="Generate text using Google's Gemini models."
)

iface.launch(share=False, debug=True)
```
